## 기계학습 과제 1 0314

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression



광고비 = np.array([1.2, 0.8, 1.0, 1.3, 0.7, 0.8, 1.0, 0.6, 0.9, 1.1, 0.6, 1.3]).reshape(-1, 1)
매출액 = np.array([101, 92, 110, 120, 90, 82, 93, 75, 91, 105, 75, 120])


모델 = LinearRegression()


모델.fit(광고비, 매출액)


plt.scatter(광고비, 매출액, color='blue', label='데이터 포인트')  # 데이터 포인트
plt.plot(광고비, 모델.predict(광고비), color='red', linewidth=2, label='선형 회귀')  # 선형 회귀 그래프
plt.title('광고비에 따른 매출액 선형 회귀 분석')
plt.xlabel('광고비')
plt.ylabel('매출액')
plt.legend()
plt.grid(True)
plt.show()
```
